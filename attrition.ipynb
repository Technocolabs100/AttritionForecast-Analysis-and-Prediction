#### Outliers detection using IQR

```python
# Calculate IQR and identify outliers with their values and counts

def find_outliers_with_values(df):
    outliers_dict = {}
    for col in df.select_dtypes(include=['int64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Find outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        if not outliers.empty:
            outliers_dict[col] = outliers
    return outliers_dict

# Find columns with outliers and their values
outliers_dict = find_outliers_with_values(df)

# Display outliers and their counts
for col, outliers in outliers_dict.items():
    print(f"Column: {col}")
    print(f"Outliers:\n{outliers}")
    print(f"Number of Outliers: {len(outliers)}\n")
```

##### Correlation matrix and HeatMap

```python
# Calculate the correlation matrix
correlation_matrix = df_encoded.corr()

# Plot the heatmap
plt.figure(figsize=(18, 12))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

#### Getting highly correlated features

```python
# Define a threshold for high correlation
threshold = 0.6

# Find features that are highly correlated
high_corr_features = set()

# Iterate through the correlation matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname1 = correlation_matrix.columns[i]
            colname2 = correlation_matrix.columns[j]
            high_corr_features.add(colname1)
            high_corr_features.add(colname2)

# Convert the set to a list 
high_corr_features = list(high_corr_features)

# Print the highly correlated features
print("Highly correlated features (correlation coefficient > 0.6):")
for feature in high_corr_features:
    print(feature)
```

#### Feature Scaling

```python
from sklearn.preprocessing import RobustScaler

# Separate features and target
features = df_encoded.drop('Attrition_encoded', axis=1)
target = df_encoded['Attrition_encoded']

# Standardize the data
scaler = RobustScaler()
scaled_data = scaler.fit_transform(features)
```

#### Principal Component Analysis 

```python
from sklearn.decomposition import PCA

# Apply PCA with 90% explained variance
pca = PCA(n_components=0.9)
principal_components = pca.fit_transform(scaled_data)

# Create DataFrame with principal components
df_pca = pd.DataFrame(data=principal_components, columns=[f'PC{i}' for i in range(1, pca.n_components_ + 1)])

# Add target variable for visualization
df_pca = pd.concat([df_pca, target.reset_index(drop=True)], axis=1)
```

```python
df_pca.info()
```

#### PCA Visualizations

```python
# Plot PCA results
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Attrition_encoded', data=df_pca, palette='Set1')
plt.title('PCA Result')
plt.show()

# Explained variance
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by components: {explained_variance}')
```

#### Identify important features through PCA

```python
# Identify important features by examining loadings for each PC
pc_loadings = pca.components_
important_features = []

for pc_idx, pc in enumerate(pc_loadings):
    top_features_idx = pc.argsort()[-3:][::-1]  
    top_features = features.columns[top_features_idx]
    important_features.extend(top_features)

important_features = list(set(important_features))

print(important_features)
```

#### SMOTE for imbalance target feature

```python
from imblearn.over_sampling import SMOTE

# Separate features and target
X = df_pca
y = df_pca['Attrition_encoded']

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Verify shapes after SMOTE
print(X_resampled.shape, y_resampled.shape)
```

#### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

#### Classification Models

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = rf_model.predict(X_test)
print(f'Model Accuracy: {accuracy_score(y_test, y_pred)}')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

```

#### Logistic Regression model

```python
# Logistic Regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Evaluate model performance
accuracy = model.score(X_test, y_test)
print(f'Model Accuracy: {accuracy}')
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#### Results Interpretation

- **Precision**: 1.00 for both classes - The ratio of true positives to the sum of true positives and false positives. Perfect precision means no false positives.
- **Recall**: 1.00 for both classes - The ratio of true positives to the sum of true positives and false negatives. Perfect recall means no false negatives.
- **F1-Score**: 1.00 for both classes - The harmonic mean of precision and recall, providing a single metric that balances both.
- **Support**: The number of actual occurrences of each class in the test set.
- **Macro Average**: Average performance across classes without considering class imbalance. Here it is 1.00 for precision, recall, and f1-score.
- **Weighted Average**: Average performance across classes, weighted by the number of instances per class. Also 1.00 in this case.

#### Implications and Considerations

- **Perfect Performance**: Both models (Logistic Regression and Random Forest) achieved perfect classification performance on the test set. This is unusual and suggests either the data is relatively easy to classify, or there may be an issue with overfitting or data leakage.

- **Possible Overfitting**:
- **Small Dataset**: If the test set is small or not representative of the broader data distribution, the model may have memorized the test data rather than generalized well.
- **Data Leakage**: Ensure that there is no leakage of test data into the training process. Data leakage can lead to inflated performance metrics.

- **Model Generalization**: To ensure that the model generalizes well to unseen data, it’s important to test on a completely independent dataset. Cross-validation can also provide a more robust estimate of model performance.

- **Validation**: If you only have a small dataset, consider techniques like k-fold cross-validation to evaluate the model more reliably.

#### Summary

The results you’re seeing indicate that both Logistic Regression and Random Forest models performed exceptionally well on the test data. However, it’s important to carefully check for issues like overfitting and data leakage and validate the model on independent data to confirm its generalizability.
