## Models training

I decided to train these 06 classifier models and I'll choose the best according to their accuracy and precision:

 SGDClassifier(),
 LinearSVC(),
 SVC(),
 KNeighborsClassifier(),
 AdaBoostClassifier(),
 RandomForestClassifier()

```python
X_scaled= scaler.fit_transform(X_selected)
X_scaled
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score,GridSearchCV

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier

# Initialiser les modèles

X_train,X_test, y_train,y_test= train_test_split(X_scaled,y, test_size=0.3, random_state=42)
models = {
    'SGDClassifier': SGDClassifier(),
    'LinearSVC': LinearSVC(),
    'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'RandomForestClassifier': RandomForestClassifier()
}

# Entraîner et évaluer les modèles
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=50)
    results[name] = scores
    print(f'{name} - Scores: {scores}')
    print(f'{name} - Mean Accuracy: {scores.mean()}')
    print("===============================================================================")
    
print("\n\n")   
    
# Afficher les résultats
for name, scores in results.items():
    print(f'{name} - Mean Accuracy: {scores.mean():.2f} - Std Deviation: {scores.std():.2f}')
    print("==================================================================================")
```

## Le tuning des parametres 

```python
import warnings
warnings.filterwarnings("ignore")
# Grille of hyperparameters
param_grids = {
    'SGDClassifier': {
        'loss': ['hinge', 'log_loss', 'modified_huber'],
        'penalty': ['l2', 'l1', 'elasticnet'],
        'alpha': [0.0001, 0.001, 0.01, 0.1]
#     },
#     'LinearSVC': {
#         'C': [0.01, 0.1, 1, 10, 100],
#         'loss': ['hinge', 'squared_hinge']
  },
#     'SVC': {
#         'C': [0.01, 0.1, 1, 10, 100],
#         'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
#         'gamma': ['scale', 'auto']
#     },
    'KNeighborsClassifier': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'AdaBoostClassifier': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.001, 0.01, 0.1, 1]
    },
    'RandomForestClassifier': {
        'n_estimators': [50, 100, 200],
        'max_features': [None, 'sqrt', 'log2'],
        'max_depth': [None, 10, 20, 30]
    }
}

# Initialize models
models = {
    'SGDClassifier': SGDClassifier(),
#     'LinearSVC': LinearSVC(),
#     'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'AdaBoostClassifier': AdaBoostClassifier(),
    'RandomForestClassifier': RandomForestClassifier()
}

# Function for grid
def tune_model(model, param_grid, X, y):
    grid_search = GridSearchCV(model, param_grid, cv=50, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X, y)
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_

# Entraîner et tuner les modèles
best_estimators = {}
for name, model in models.items():
    print(f'Tuning {name}...')
    best_estimator, best_params, best_score = tune_model(model, param_grids[name], X_train, y_train)
    best_estimators[name] = best_estimator
    print(f'Best params for {name}: {best_params}')
    print(f'Best score for {name}: {best_score}')

# Résumé des meilleurs modèles
for name, estimator in best_estimators.items():
    print(f'Best estimator for {name}: {estimator}')
```

# Best model with best parameters

Model: RandomForestClassifier 
Params: max_depth:30; max_features: 'sqrt'; n_estimators:10

with a score: 0.84

```python
### best model 
best_model = RandomForestClassifier(max_depth=30, random_state=42, max_features='log2', n_estimators=200)
best_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# make prediction with X_test
y_pred = best_model.predict(X_test)

# Calulate precision
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")


# display confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```
